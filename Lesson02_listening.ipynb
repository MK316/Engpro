# Setting

```python
#@markdown üîÑ Making functions: etts('text'), ktts('text')
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = True)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)
  
  gtts_object.save("K-audio.mp3")
  return Audio("K-audio.mp3")
```

# üçÖ A. You'll be hearing three English words, one of which is different from the others. Choose the one that is pronounced differently.

```python
#@markdown üéß Instruction: You'll be hearing three English words, one of which is different from the others. Choose the one that is pronounced differently

mytxt = "You'll be hearing three English words, one of which is different from the others. Choose the one that is pronounced differently."
etts(mytxt)
Audio("E-audio.mp3")
```

```python
#@markdown üéß Example: "mitt, meat, meat"

mytxt = "mitt, meat, meat. Which one is different?"
etts(mytxt)
Audio("E-audio.mp3")
```
